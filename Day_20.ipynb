```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# The input text for generating the WordCloud.
text = 'Artificial Intelligence Machine Learning'

# Create a WordCloud object with custom parameters.
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Save the WordCloud image to a file.
wordcloud.to_file('wordcloud.png')
print("WordCloud image saved as 'wordcloud.png'.")

# Optional: Display the generated WordCloud.
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud Visualization')
plt.show()
```
