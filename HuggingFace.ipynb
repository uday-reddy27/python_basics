```python
from transformers import BertTokenizer, TFAutoModelForSequenceClassification
import tensorflow as tf
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-multilingual-cased")
```

```python
text = "I love learning about NLP!"
tokenizer(text)
```

```python
print("Tokenized Words:", tokenizer.tokenize(text))
tokens = tokenizer(text)
print("Token IDs:", tokens["input_ids"])
```

```python
outputs = model(**tokens)
```

```python
generator = pipeline("text-generation", model="ProsusAI/finbert")

prompt = "There once lived a king"
result = generator(prompt, max_length=100, num_return_sequences=1)

print(result)
```

```python
qa_pipeline = pipeline("question-answering",model='deepset/roberta-base-squad2')

context = """Virat Kohli is an Indian international cricketer who plays Test and ODI cricket for the national team and is a former captain in all formats. He is a right-handed batsman and an occasional right arm medium pace bowler. Wikipedia
Born: 5 November 1988 (age 36 years), Delhi
Spouse: Anushka Sharma (m. 2017)
Dates joined: 18 August 2008 (India national cricket team) · See more
Current teams: India national cricket team, Royal Challengers Bengaluru (Batter)
Number of centuries: 80
Height: 1.75 m
Siblings: Bhawna Kohli Dhingra, Vikas Kohli
"""

question = "Who is Virat Kohli?"

result = qa_pipeline(question=question, context=context)

print(result)
```
